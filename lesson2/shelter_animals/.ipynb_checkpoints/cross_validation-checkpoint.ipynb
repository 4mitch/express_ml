```python
df['SexuponOutcome'] = df['SexuponOutcome'].fillna('Unknown')
df_test['SexuponOutcome'] = df_test['SexuponOutcome'].fillna('Unknown')
```

```python
df['SexuponOutcome'].value_counts(dropna=False)
```

```python
df['SexuponOutcome'] = df['SexuponOutcome'].astype('str')
df_test['SexuponOutcome'] = df_test['SexuponOutcome'].astype('str')
```

```python
def detect_sterilization(x):
    if 'Intact' in x:
        return 'Intact'
    if 'Spayed' in x or 'Neutered' in x:
        return 'Neutered'
    return 'Unknown'
def detect_gender(x):
    if 'Male' in x:
        return 'Male'
    if 'Female' in x:
        return 'Female'
    return 'Unknown'
df['Sterilization'] = df['SexuponOutcome'].map(detect_sterilization)
df['Gender'] = df['SexuponOutcome'].map(detect_gender)

df_test['Sterilization'] = df['SexuponOutcome'].map(detect_sterilization)
df_test['Gender'] = df['SexuponOutcome'].map(detect_gender)
```

```python
df['Sterilization'].value_counts()
```

```python
df['Gender'].value_counts()
```

```python
sterilization = pd.get_dummies(
    df['Sterilization'], prefix='Sterilization', drop_first=True)
df = pd.concat([df, sterilization], axis=1)
gender = pd.get_dummies(df['Gender'], prefix='Gender', drop_first=True)
df = pd.concat([df, gender], axis=1)

sterilization = pd.get_dummies(df_test['Sterilization'], prefix='Sterilization', drop_first=True)
df_test = pd.concat([df_test, sterilization], axis=1)
gender = pd.get_dummies(df_test['Gender'], prefix='Gender', drop_first=True)
df_test = pd.concat([df_test, gender], axis=1)
```

```python
df['AgeuponOutcome'] = df['AgeuponOutcome'].fillna('unknown')
df_test['AgeuponOutcome'] = df_test['AgeuponOutcome'].fillna('unknown')
```

```python
def get_age_in_weeks(age_str):
    if age_str == 'unknown':
        return -50
    number, time_period = age_str.split()
    multiplicator = 1
    if time_period == 'years':
        multiplicator = 365 / 7
    elif time_period == 'months':
        multiplicator = 30 / 7
    return int(number) * multiplicator

df['AgeuponOutcome'] = df['AgeuponOutcome'].map(get_age_in_weeks)
df_test['AgeuponOutcome'] = df_test['AgeuponOutcome'].map(get_age_in_weeks)
```

Найдем оптимальное количество деревьев для случайного леса по кросс-валидации:

```python
kfold = StratifiedKFold(n_splits=3, random_state=42, shuffle=True)

cv_scores = []

estimators_counter = xrange(1, 601, 10)
for n in estimators_counter:
#     print n
    clf = RandomForestClassifier(random_state=42, n_estimators=n, n_jobs=8)
    cv_score = cross_val_score(clf, X, y, cv=kfold)
    cv_scores.append(cv_score)
```

```python
figure(figsize=(10,5))
cv_scores = np.array(cv_scores)
    
scores_mean = np.mean(cv_scores, axis=1)
scores_std = np.std(cv_scores, axis=1)
plt.grid()
plt.fill_between(estimators_counter, 
                 scores_mean - scores_std, 
                 scores_mean + scores_std, 
                 alpha=0.1, color='g')
plt.plot(estimators_counter, scores_mean, 'o-', color='g', label='test score')

plt.xlabel('n_estimators')
plt.ylabel('Score')
_ = plt.legend(loc='lower right')
```

#### Как выбрать количество фолдов?

```python
%precision 4
from sklearn.model_selection import StratifiedKFold, cross_val_score

cv_scores = []
for n_splits in xrange(2, 20):
    cv_scores.append([])
    print 'n_splits={}'.format(n_splits)
    for i in xrange(20):
        kfold = StratifiedKFold(n_splits=n_splits, random_state=i, shuffle=True)
        cv_score = cross_val_score(RandomForestClassifier(random_state=42, n_estimators=10),
                                   df[training_columns], df['OutcomeType'], cv=kfold)
#         print 'i={}, mean={}, std={}'.format(i, mean(cv_score), std(cv_score))
        cv_scores[-1].append(mean(cv_score))
```
